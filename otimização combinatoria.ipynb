# Gráfico de Dispersão dos Valores de Abertura

```python
import matplotlib.pyplot as plt

data = {
    "Data": ["02/10/2023", "03/10/2023", "04/10/2023", "05/10/2023", "06/10/2023"],
    "Valor Abertura": [132.15, 134.93, 133.66, 136.13, 134.94]
}

x = range(1, len(data["Data"]) + 1)  
y = data["Valor Abertura"]


plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", label="Valor Abertura")
plt.title("Gráfico de Dispersão dos Valores de Abertura", fontsize=14)
plt.xlabel("Índice da Data", fontsize=12)
plt.ylabel("Valor de Abertura", fontsize=12)
plt.xticks(ticks=x, labels=data["Data"], rotation=45)
plt.ylim(130, 140)
plt.xlim(0, 6)
plt.grid(alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

# Gráfico de Dispersão com Reta de Regressão L2

```python

alpha = 132.328
beta = 0.678

regression_line = [alpha + beta * xi for xi in x]

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", label="Valor Abertura")
plt.plot(x, regression_line, color="red", label="Reta de Regressão (L2)", linestyle="--")
plt.title("Gráfico de Dispersão com Reta de Regressão L2", fontsize=14)
plt.xlabel("Índice da Data", fontsize=12)
plt.ylabel("Valor de Abertura", fontsize=12)
plt.xticks(ticks=x, labels=data["Data"], rotation=45)
plt.ylim(130, 140)
plt.xlim(0, 6)
plt.grid(alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

# Gráfico de Dispersão com Reta de Regressão L1

```python

alpha = 131.395
beta = 0.755

regression_line = [alpha + beta * xi for xi in x]

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", label="Valor Abertura")
plt.plot(x, regression_line, color="red", label="Reta de Regressão (L1)", linestyle="--")
plt.title("Gráfico de Dispersão com Reta de Regressão L1", fontsize=14)
plt.xlabel("Índice da Data", fontsize=12)
plt.ylabel("Valor de Abertura", fontsize=12)
plt.xticks(ticks=x, labels=data["Data"], rotation=45)
plt.ylim(130, 140)
plt.xlim(0, 6)
plt.grid(alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

# Gráfico de Dispersão com as Retas de Regressão L1 eL2

```python
alpha_l1 = 131.395
beta_l1 = 0.755

alpha_l2 = 132.328
beta_l2 = 0.678


regression_line_l1 = [alpha_l1 + beta_l1 * xi for xi in x]
regression_line_l2 = [alpha_l2 + beta_l2 * xi for xi in x]

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", label="Valor Abertura")
plt.plot(x, regression_line_l1, color="green", label="Reta de Regressão (L1)", linestyle="-")
plt.plot(x, regression_line_l2, color="red", label="Reta de Regressão (L2)", linestyle="--")
plt.title("Gráfico de Dispersão com as Retas de Regressão L1 e L2", fontsize=14)
plt.xlabel("Índice da Data", fontsize=12)
plt.ylabel("Valor de Abertura", fontsize=12)
plt.xticks(ticks=x, labels=data["Data"], rotation=45)
plt.ylim(130, 140)
plt.xlim(0, 6)
plt.grid(alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```
